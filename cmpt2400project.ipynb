```python
import pandas as pd

# Load the Excel file (Make sure the file is in the same directory as your script)
file_name = "NPRI_2000-2022.xlsx"

# Load the specific sheet for disposal and treatment data
df = pd.read_excel(file_name, sheet_name="Disposals and Transfers 00-20")

# Display the first few rows
df.head()
```

```python
import pandas as pd

# Convert "Number of employees" to numeric and replace errors with NaN
df["Number of employees"] = pd.to_numeric(df["Number of employees"], errors="coerce")

# Replace 0 values with NaN (avoiding inplace=True)
df["Number of employees"] = df["Number of employees"].replace(0, pd.NA)

# Fill missing values with NAICS industry median
df["Number of employees"] = df.groupby("NAICS / Code_SCIAN")["Number of employees"].transform(lambda x: x.fillna(x.median()))

# Fill remaining missing values with global median (avoiding inplace=True)
df["Number of employees"] = df["Number of employees"].fillna(df["Number of employees"].median())

# Check missing values after fixing
print(f"Missing values in 'Number of employees': {df['Number of employees'].isnull().sum()}")
```

```python
# Reload the dataset since execution state was reset




# Step 1: Identify Missing Values
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

# Create a dataframe to display missing values and their percentage
df_missing = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})

# Step 2: Handle Missing Values
# Options:
# - Fill numerical missing values with the median of the respective column
# - Fill categorical missing values with the mode (most frequent value)

# Separate numerical and categorical columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Fill missing values
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].median())


# Display the missing values before and after cleaning
# Instead of using ace_tools, display the DataFrame using pandas' display function
print("Missing Values Before Cleaning:")
display(df_missing)

# Verify missing values after cleaning
missing_after_cleaning = df.isnull().sum()
df_missing_after = pd.DataFrame({'Missing Values After Cleaning': missing_after_cleaning})

# Display the missing values after cleaning using pandas' display function
print("\nMissing Values After Cleaning:")
display(df_missing_after)
```

```python
import pandas as pd


# Convert all values to lowercase strings
df["Facility_Name / Installation"] = df["Facility_Name / Installation"].astype(str).str.lower()

# Replace placeholders with NaN
placeholders = ["(blank)", "-", "<na>", "(same)", "same", "same as above"]
#df["Facility_Name / Installation"].replace(placeholders, pd.NA, inplace=True)
df.replace({"Facility_Name / Installation":placeholders}, inplace=True)
#df.method({col: value}, inplace=True)
# Count missing values after cleaning
print(f"Missing Facility Names after standardization: {df['Facility_Name / Installation'].isnull().sum()}")

# Count unique facilities per company
facility_count = df.groupby("Company_Name / Dénomination_sociale_de_l'entreprise")["Facility_Name / Installation"].nunique()

# Find the most frequently reported facility for each company
most_common_facility = df.groupby("Company_Name / Dénomination_sociale_de_l'entreprise")["Facility_Name / Installation"].agg(lambda x: x.mode()[0] if not x.mode().empty else pd.NA)

# Display how many companies have multiple facilities
print(f"Companies with multiple facilities: {(facility_count > 1).sum()}")

# Function to impute missing Facility Names
def impute_facility_name(row):
    if pd.isna(row["Facility_Name / Installation"]):
        company = row["Company_Name / Dénomination_sociale_de_l'entreprise"]

        # If the company has only one facility, use the company name as the facility name
        if facility_count[company] == 1:
            return company

        # If it has multiple facilities, use the most frequent one
        return most_common_facility[company]

    return row["Facility_Name / Installation"]

# Apply imputation function
df["Facility_Name / Installation"] = df.apply(impute_facility_name, axis=1)

# Check remaining missing values
print(f"Missing values after imputation: {df['Facility_Name / Installation'].isnull().sum()}")
```

```python
import pandas as pd

### 🔹 1. Convert City Names to Lowercase ###
df["City"] = df["City"].str.lower()

# Count missing values
print(f"Missing 'City' values before fixing: {df['City'].isna().sum()}")

### 🔹 2. Fill Missing City Using Latitude & Longitude ###
# Create a mapping of (Latitude, Longitude) to the most common City
lat_lon_to_city = df.groupby(["Latitude", "Longitude"])["City"].agg(lambda x: x.mode()[0] if not x.mode().empty else pd.NA)

# Assign city based on matching Latitude & Longitude
df["City"] = df.apply(lambda row: lat_lon_to_city.get((row["Latitude"], row["Longitude"]), row["City"]), axis=1)

# Check remaining missing values
print(f"Missing 'City' values after Lat/Lon fill: {df['City'].isnull().sum()}")

### 🔹 3. Fill Remaining Missing City Using Province ###
# Get the most common city per province
most_common_city_by_province = df.groupby("PROVINCE")["City"].agg(lambda x: x.mode()[0] if not x.mode().empty else "Unknown City")

# Assign city based on province
df["City"] = df.apply(lambda row: most_common_city_by_province[row["PROVINCE"]] if pd.isnull(row["City"]) else row["City"], axis=1)

# Final check for missing values
print(f"Final missing 'City' values: {df['City'].isnull().sum()}")

```
